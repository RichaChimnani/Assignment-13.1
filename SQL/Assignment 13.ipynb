```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
url= "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
df = pd.read_csv(url) 
df.head()
```

```python
df.columns = ["age","workclass","fnlwgt","education","education_num","marital_status","occupation","relationship","race",
             "sex","capital_gain","capital_loss","hours_per_week","native_country","income_class"]
df.head()
```

```python
df.shape
```

```python
from sqlalchemy import create_engine
from pandasql import sqldf

engine = create_engine("sqlite:///sqladb", echo= False) # Creating engine

df.to_sql("sqladb", engine, if_exists='replace')
```

```python
conn  = sqlite3.connect('sqladb')
```

```python
sqladb = df
```

#1. Select 10 records from the adult sqladb

```python
print("\n 1. Select 10 records from the adult sqladb\n",'-'*50, sep='')

sqldf('SELECT * FROM sqladb limit 10;') 
```

#2. Show me the average hours per week of all men who are working in private sector

```python
print("\n 2. Show me the average hours per week of all men who are working in private sector\n",'-'*100, sep='')

sqldf ('SELECT avg(hours_per_week) as [Average Hrs/week] FROM sqladb Group BY workclass="Private";')
```

#3. Show me the frequency table for education, occupation and relationship, separately

```python
print('''\n 3. Show me the frequency table for education, occupation and relationship, separately 
       (Education Frequency Table)\n''','-'*100, sep='')

sqldf('SELECT education, COUNT(*) AS Frequency FROM sqladb GROUP BY education;')
```

```python
print('''\n 3. Show me the frequency table for education, occupation and relationship, separately 
       (Occupation Frequency Table)\n''','-'*100, sep='')

sqldf('SELECT occupation, COUNT(*) AS Frequency FROM sqladb GROUP BY occupation;')
```

```python
print('''\n 3. Show me the frequency table for education, occupation and relationship, separately 
       (Relationship Frequency Table)\n''','-'*100, sep='')

sqldf('SELECT relationship, COUNT(*) AS Frequency FROM sqladb GROUP BY relationship;')
```

#4. Are there any people who are married, working in private sector and having a masters degree

```python
print("\n 4. Are there any people who are married, working in private sector and having a masters degree \n",'-'*100, sep='')

sqldf ('SELECT count(*) FROM sqladb WHERE education=" Masters" AND workclass=" Private" AND marital_status=" Married-civ-spouse";')
```

#5. What is the average, minimum and maximum age group for people working in different sectors

```python
print("\n 5. What is the average, minimum and maximum age group for people working in different sectors \n",'-'*100, sep='')

sqldf ("SELECT occupation, Min(age), Max(age), Avg(age) FROM sqladb Group By occupation;")
```

#6. Calculate age distribution by country

```python
print("\n 6. Calculate age distribution by country \n",'-'*60, sep='')

sqldf('SELECT native_country,Min(age),Max(age),Avg(Age) FROM sqladb Group By native_country;')
```

#7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

```python
print("\n 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss' \n",'-'*100, sep='')

sqladb['Net_Capital_Gain'] = sqladb['capital_gain'] - sqladb['capital_loss']

sqldf('Select * FROM sqladb Limit 5;')
```
