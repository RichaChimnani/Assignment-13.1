```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data")
df
```

```python
df.columns = ["age","workclass","fnlwgt","education","education_num","marital_status","occupation","relationship","race",
             "sex","capital_gain","capital_loss","hours_per_week","native_country","income_class"]
df.head()
```

#1. Create an sqlalchemy engine using a sample from the data set

```python

from sqlalchemy import create_engine
from pandasql import sqldf

engine = create_engine('sqlite:///:memory:', echo=False)  # Creating engine

df.to_sql("sqladb", engine, if_exists='replace')
```

```python
connection  = sqlite3.connect('sqladb')

cursor = connection.cursor()
```

```python
print("\n 1. Create an sqlalchemy engine using a sample from the data set\n",'-'*80, sep='')

print(pd.read_sql_query("SELECT * FROM sqladb Limit 5", connection));
```

#2. Write two basic update queries

```python
print("\n 2. First Update (Changing Private workclass to Updated)\n",'-'*80, sep='')

cursor.execute('UPDATE sqladb SET workclass = "Updated" WHERE workclass = " Private"')
print(pd.read_sql_query("SELECT * FROM sqladb", connection));
```

```python
print("\n 2. Second Update (Changing native-country from United-States to America)\n",'-'*80, sep='')

cursor.execute('UPDATE sqladb SET native_country = "America" WHERE native_country = " United-States"')
print(pd.read_sql_query("SELECT * FROM sqladb;", connection))
```

#3. Write two delete queries

```python
print("\n 3. First delete query (Delete rows where education = Masters) \n",'-'*80, sep='')

cursor.execute('DELETE FROM sqladb WHERE education = " Masters"')
print(pd.read_sql_query("SELECT * FROM sqladb", connection))
```

```python
print("\n 3. Second delete query (Delete rows where sex = Male) \n",'-'*80, sep='')

cursor.execute('DELETE FROM sqladb WHERE sex = " Male"')
print(pd.read_sql_query("SELECT * FROM sqladb", connection))
```

#4. Write two filter queries

```python
print('\n 4. Write two filter queries \n','-'*75, sep='')

print(pd.read_sql_query("SELECT education, COUNT(*) AS Total FROM sqladb GROUP BY education;", connection))
```

```python
print('\n 4. Write two filter queries \n','-'*35, sep='')

print(pd.read_sql_query("SELECT race, COUNT(*) AS Total FROM sqladb GROUP BY race;", connection))
```

#5. Write two function queries

```python
print("\n 5. First function query (Working hours as per Occupation) \n",'-'*65, sep='')

def workhrs():
    
    connection  = sqlite3.connect("sqladb")
    cursor = connection.cursor()
           
    return pd.read_sql_query('SELECT occupation, hours_per_week FROM sqladb GROUP BY occupation ', connection)

workhrs()
```

```python
print("\n 5. Second function query (Average age for getting married) \n",'-'*65, sep='')

def maritalage():
    
    connection  = sqlite3.connect("sqladb")
    cursor = connection.cursor()
    
    return pd.read_sql_query('SELECT marital_status, avg(age) FROM sqladb GROUP BY marital_status ', connection)

maritalage()
```
